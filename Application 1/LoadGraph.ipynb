```python
def  make_complete_graph(num_nodes):
    complete_graph={}
    for  node in range(num_nodes):
        complete_graph[node]=set(range(num_nodes))-{node}
    return complete_graph
```

```python
#general imports
import  random

class DPATrial :
    """
    Simple   class  to  encapsulate  trials  for DPA  alghoritm .
    Maintains  a list of  node  numbers with multiple instances  for   each number.
    The  number  of instances of each node  number  are in the same  proportion  as  the desired probabilities
    Using  random.choice()  to select a node  number  from the  list  for  each trials .
    """
    def init(self,num_nodes):
        """
        Initialize  a  DPATrial  object corresponding to  complete  graph with num_nodes  nodes
        Note that  the initial  list of   node numbers  has  num_nodes  copies  of  each  node numbe
        """
        self.num_nodes=num_nodes
        self.node_numbers=[node  for  node  in range(num_nodes) for  dummy_idx  in  range(num_nodes)]
    def run_trail(self,num_nodes):
        """
        Conduct  num_node trials   using by applyng  random.choice() to the list of   node  numbers
        Updated the list of  node numbers   such  that  each node number is  in the same  ratio as  the desired 
        probabilities.
        Returns :
        Set of nodes
        """
        #compute  the neighbor  for the newly created  node :
        new_node_neighbors=set()
        for  dummy_ix  in  range(num_nodes):
            new_node_neighbors.add(random.choice(self.node_numbers))
        #update the list of  node numbers  such   that  each number  appears in the correct  ratio
        self.node_numbers.append(self.num_nodes)
        self.node_numbers.extend(list(new_node_neighbors))
        #update  the  number of  nodes
        self.num_nodes+=1
        return  new_node_neighbors
```

```python
def dpa(num_of_nodes, m):
    import itertools
    assert 1 <= m <= num_of_nodes
    graph = make_complete_graph(m)
    indeg = compute_in_degrees(graph)
    nodes_ps = list(itertools.chain.from_iterable([node] * (indeg[node] + 1) for node in graph.keys()))
    for i in range(m, num_of_nodes):
        nodes = set()
        for _ in range(m):
            node = random.choice(nodes_ps)
            nodes.add(node)
        graph[i] = set()
        nodes_ps += list(nodes)
        nodes_ps.append(i)
        for node in nodes:
            graph[i].add(node)
    return graph
```
